```python
plt.figure(figsize=(15,15))
col=["r","y","c","olive","orange","peru","b","g"]
plotnumber=1
j=0
for i in toyota_df:
    if plotnumber<=9:
        ax=plt.subplot(3,3,plotnumber)
        plt.scatter(y=toyota_df["Price"],x=toyota_df[i],color=col[j])
        plt.xlabel(i,fontsize=10)
        plt.ylabel("Price",fontsize=10)
        plotnumber+=1
        j+=1
```

```python
import statsmodels.formula.api as smf
model=smf.ols("Price ~ Age_08_04 + KM + HP + cc + Doors + Gears + Quarterly_Tax + Weight",data=toyota_df).fit()
```

```python
model.params
```

```python
print(model.tvalues,'\n',model.pvalues)
```

```python
model.rsquared, model.rsquared_adj
```

```python
rsq_Age_08_04 = smf.ols('Age_08_04 ~ KM + HP + cc + Doors + Gears + Quarterly_Tax + Weight',data=toyota_df).fit().rsquared  
vif_Age_08_04 = 1/(1-rsq_Age_08_04) 

rsq_KM = smf.ols('KM ~ Age_08_04 + HP + cc + Doors + Gears + Quarterly_Tax + Weight',data=toyota_df).fit().rsquared  
vif_KM = 1/(1-rsq_KM) 

rsq_HP = smf.ols('HP ~ Age_08_04 + KM + cc + Doors + Gears + Quarterly_Tax + Weight',data=toyota_df).fit().rsquared  
vif_HP = 1/(1-rsq_HP) 

rsq_cc = smf.ols('cc ~ Age_08_04 + KM + HP + Doors + Gears + Quarterly_Tax + Weight',data=toyota_df).fit().rsquared  
vif_cc = 1/(1-rsq_cc) 

rsq_Doors = smf.ols('Doors ~ Age_08_04 + KM + HP + cc + Gears + Quarterly_Tax + Weight',data=toyota_df).fit().rsquared  
vif_Doors = 1/(1-rsq_Doors) 

rsq_Gears = smf.ols('Gears ~ Age_08_04 + HP + cc + Doors + Quarterly_Tax + Weight',data=toyota_df).fit().rsquared  
vif_Gears = 1/(1-rsq_Gears) 

rsq_Quarterly_Tax = smf.ols('Quarterly_Tax  ~ Age_08_04 + KM + cc + Doors + Gears + Weight',data=toyota_df).fit().rsquared  
vif_Quarterly_Tax = 1/(1-rsq_Quarterly_Tax) 

rsq_Weight = smf.ols(' Weight ~ Age_08_04 + KM + HP + Doors + Gears + Quarterly_Tax',data=toyota_df).fit().rsquared  
vif_Weight = 1/(1-rsq_Weight) 
# Storing vif values in a data frame
d1 = {'Variables':['Age_08_04', 'KM', 'HP', 'cc', 'Doors', 'Gears','Quarterly_Tax', 'Weight'],'VIF':[vif_Age_08_04,vif_KM,vif_HP,vif_cc,vif_Doors,vif_Gears,vif_Quarterly_Tax,vif_Weight]}
Vif_frame = pd.DataFrame(d1)  
Vif_frame
```
